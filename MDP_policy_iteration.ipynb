```python
import numpy as np


states = [0,1,2,3,4]
actions = [0,1]

P = np.zeros((5, 2, 5))  
R = np.zeros((5, 2, 5))  

P[0,0,1] = 1.0
P[1,1,2] = 1.0
P[2,0,3] = 1.0
P[3,1,4] = 1.0
P[4,0,4] = 1.0


R[0,0,1] = 1
R[1,1,2] = 10
R[2,0,3] = 50
R[3,1,4] = 20
R[4,0,4] = 5


gamma = 0.9

policy = [0 for s in range(5)]
V = np.zeros(5)

print("Initial policy", policy)


is_value_changed = True
iterations = 0
while is_value_changed:
    is_value_changed = False
    iterations += 1
    for s in range(5):
        V[s] = sum([P[s,policy[s],s1] * (R[s,policy[s],s1] + gamma*V[s1]) for s1 in range(5)])

    for s in range(5):
        q_best = V[s]
        for a in range(2):
            q_sa = sum([P[s, a, s1] * (R[s, a, s1] + gamma * V[s1]) for s1 in range(5)])
            if q_sa > q_best:
                print("State", s, ": q_selected_action", q_sa, "q_best_action", q_best)
                policy[s] = a
                q_best = q_sa
                is_value_changed = True

    print("Iterations:", iterations)

print("Optimal policy")
print(policy)
print(V)
```
